Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

            - To ensure equal representation of both genders, the correct method is Stratified Sampling.

Answer

You should use Stratified Sampling.

Why?

The population has 60% boys and 40% girls.

In Stratified Sampling, the population is divided into groups (strata) such as boys and girls.

Then samples are taken separately from each group, ensuring that both genders are properly represented.

Why not Simple Random Sampling?

In Simple Random Sampling, every student has an equal chance of being selected, but:

The sample might accidentally contain more boys or more girls.

It does not guarantee equal representation.

✅ Final Answer:
Use Stratified Sampling because it divides the population into boys and girls (strata) and allows you to select samples from each group to ensure equal gender representation.

Question 10: Differentiate between discrete and continuous random variables with examples.

           - 1️⃣ Discrete Random Variables

Definition:
A discrete random variable can take countable, distinct values (often integers).

The values can be listed, and there are gaps between them.

Usually comes from counting something.

Examples:

Number of students in a class → 0, 1, 2, …

Number of defective bulbs in a batch → 0, 1, 2, …

Number of heads in 3 coin tosses → 0, 1, 2, 3

Key Points:

Values are finite or countably infinite

Probability of each value can be listed in a probability mass function (PMF)

2️⃣ Continuous Random Variables

Definition:
A continuous random variable can take any value within an interval.

Comes from measuring something.

The variable can take infinitely many values, often decimals or fractions.

Examples:

Height of students → 150.2 cm, 158.7 cm, etc.

Time taken to run 100 m → 12.34 s, 12.345 s, etc.

Weight of a bag of rice → 5.001 kg, 5.002 kg, …

Key Points:

Values are uncountably infinite

Probability is calculated using probability density function (PDF)

Probability of any exact single value = 0; we consider intervals instead
